```python
print('과일가게')
print('회원가입')
name = input('이름을 입력해주세요')
id = input('id를 입력해주세요')
pw = input('비밀번호를 입력해주세요')
print('회원가입 성공')
print('로그인')
id1 = input('id를 입력해주세요 : ')
pw1 = input('비밀번호를 입력해주세요')
if id==id1 and pw==pw1:
    print('로그인에 성공하셨습니다')
    print(f'안녕하세요{name}님 무엇을 하시겠습니까?')
    be = input('1.입금 2.과일구매(현금)')
    if be=='1':
        nowmoney = int(input('입력하신 금액을 입급합니다.:\n원'))
        print(f'입금된 금액은 {nowmoney}원 입니다.\n무엇을 하시겠습니까?')
        be1 = input('입금 후 무엇을 하시겠습니까?\n1.잔액확인 2.과일구매')
        if be1=='1':
            print(f'현재 금액은 {nowmoney}원입니다.')
        elif be1=='2':
            print(f'어서오세요{name}님 필요하신 과일은 무엇인가요?')
            fr = input('사과, 배, 수박')
            if fr== '사과':
                frp=500
            elif fr== '배':
                frp=1000
            elif fr== '수박':
                frp=5000
            frn = int(input('개수를 입력해주세요'))
            frs = frp*frn
            print(f'{name}님 %s %d개를 구매하셨고 가격은 %d원 입니다.'%(fr,frn,frs))
            money = nowmoney - frs
            if int(money)<0:
                print('잔액이 부족합니다.')
            elif int(money)>=0:
                print(f'{name}님 현재 남은 잔액은{money}원입니다.')
        else:
            print('잘못된입력입니다.')
    elif be=='2':
        print(f'어서오세요{name}님 필요하신 과일은 무엇인가요?')
        fr = input('사과, 배, 수박')
        if fr== '사과':
            frp=500
        elif fr== '배':
            frp=1000
        elif fr== '수박':
            frp=5000
        frn = int(input('개수를 입력해주세요'))
        frs = frp*frn
        print(f'{name}님 %s %d개를 구매하셨고 가격은 %d원 입니다.'%(fr,frn,frs))

else:
    print('로그인에 실패하셨습니다')
```
